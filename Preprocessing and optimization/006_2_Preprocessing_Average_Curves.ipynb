```python
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import fdasrsf as fs
import numpy as np
import json
from pykalman import KalmanFilter
```

```python
df = scipy.io.loadmat('TremorStruct_V1_Joint_Severity_Constancy_After_Exclusion.mat')
```

```python
var = df['TremorStruct'][0]
pat_number = len(var)
pat_number
```

```python
responses = {} # dictionary of scalare responses
data = {} # dictionary of functional data
for i in range(pat_number):
    data_pat = var[i]
    data[str(i)]=data_pat[17][0,0][0][:,:]
    
    severity, constancy, age, gender, payed, activity = data_pat[1],data_pat[2],data_pat[3],data_pat[4],data_pat[5],data_pat[6]
    responses[str(i)]=list([i, severity[0][0], constancy[0][0], age[0][0], gender[0][0], payed[0][0], activity[0][0]])
```

```python
aligned = {}
for pat in data.keys():
    time = np.arange(data[pat].shape[0]).astype(np.float64)
    
    if len(time)>0:
        
    #initialize KF
        kf = KalmanFilter(transition_matrices=np.eye(data[pat].shape[1]),transition_covariance=10000*np.eye(data[pat].shape[1]),
                  observation_matrices=np.eye(data[pat].shape[1]),observation_covariance=10000*np.eye(data[pat].shape[1]))
    #fit KF
        kf = kf.em(data[pat])
        (smoothed_state_means, smoothed_state_covariances) = kf.smooth(data[pat])
    
    #alignment
        obj = fs.fdawarp(smoothed_state_means.astype(np.float64), time)
        obj.srsf_align(parallel=True)
        aligned[pat] = obj.fmean
```

```python
aligned.keys()
```

```python
dataset = pd.DataFrame()

for j in aligned.keys():
    dataset[str(j)] = np.concatenate([responses[j],aligned[j]],axis=0)
```

```python
index = ['pat_number','severity', 'constancy', 'age', 'gender', 'payed', 'activity']+list(range(288))
dataset['idx']=index
dataset.set_index('idx', inplace=True)
```

```python
dataset.to_csv('average_curves.csv',index = True, index_label='idx')
```
