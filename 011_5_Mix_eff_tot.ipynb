```python
def prior_kernel(s,t):
    return np.exp(-((t-s)/5)**2)+0.0001*(s==t)
sigma = 1
T = np.linspace(1, 288, 288)
K = np.fromfunction(np.vectorize(lambda s, t: prior_kernel(T[s], T[t])), (T.shape[0], T.shape[0]), dtype=int)
```

```python
T0=pd.read_csv("chain_fixed.csv") #from the sample with sigma=1, L=5, alpha=1
T0,_=stats.mode(T0[500:], axis=0)
T0=T0.flatten().tolist()
```

```python
def get_K0():
        tau = np.where(T0==1)[0]
        K_tau=K[:, tau]
        K_tau_tau = K[tau[:, None], tau]
        
        if np.linalg.det(K_tau_tau)==0:
            K_tau_tau = K_tau_tau+0.0001*np.eye(len(tau))
            
        return K- np.dot(np.dot(K_tau, np.linalg.inv(K_tau_tau)), K_tau.T)
K0=get_K0()+np.eye(288)*0.00001
```

## Mixed model with random effects on levodopa treatments and sex
Let the model be $$Y_{i}=\int_{T} x_i(t) \beta(t)dt+\lambda_i \int_{T}  x_i(t) \gamma(t) dt+ z_i \theta+\varepsilon_i$$
$$ \beta|\mu,\sigma^{2}, T_{0} \sim GP(0,\sigma^{2}K^{0})$$ 
$$ \gamma \sim \mathcal{N}(0, \sigma_0^{2}K^{0})$$
$$ \varepsilon_i \sim \mathcal{N} (0, \sigma^2)$$






```python
# Set sigma and sigma0 to 1
mix = """
data {
    int<lower=1> N; 
    int<lower=0> g; 
    matrix[N, g] X; 
    vector[N] Y; 
    vector[N] lev; // Group membership indicator for each observation
    vector [N] sex;
    matrix[g, g] K0;
    vector[g] mu;
    real <lower=0.01> sigma;
    real <lower=0.01> sigma0;
}

parameters {
    vector[g] beta; 
    vector[g] gamma;
    real theta;
    
}

model {
    beta ~ multi_normal(mu, sigma*K0);
    gamma ~ multi_normal(mu, sigma0*K0);  //rand eff
    theta ~normal(0,1);  //rand eff
   
    for (n in 1:N) {
        real integral_fix = dot_product(beta, X[n]);
        real integral0 = dot_product(gamma, X[n]);
        Y[n] ~ normal(integral_fix+lev[n]*integral0+sex[n]*theta, sigma); // mixed eff likelihood
    }
}



"""

# Write stan model to file
stan_file = "./stan/tot/mix.stan"
with open(stan_file, "w") as f:
    print(mix, file=f)

# Compile stan model
mix= CmdStanModel(stan_file=stan_file) #max_treedepth=15
```

```python

data = {
    "N": len(y23),
    "g": x.shape[0],
    "X": np.transpose(x),
    "Y": y23,
    "lev": lev_g,
    "sex": sex,
    "K0": K0,
    "mu": np.zeros(288),
    "sigma": 1, 
    "sigma0": 1,
}  
mix_fit = mix.sample(data=data, chains=2,parallel_chains=2, adapt_delta=0.95, iter_warmup=100, iter_sampling=500)


# Convert to arviz data type
mix_data = az.from_cmdstanpy(mix_fit)
```

```python
beta=np.array(mix_data.posterior.beta)[0]
gamma=np.array(mix_data.posterior.gamma)[0]
theta=np.array(mix_data.posterior.theta)[0]
beta_hat=np.mean(beta[300:], axis=0)
gamma_hat=np.mean(gamma[300:], axis=0)
theta_hat=np.mean(theta[300:], axis=0)
```

```python
def compute_predictions(new_data):
    predictions = []
    for i in range(136):
        integral_fix = np.dot(new_data.iloc[i],beta_hat)
        integral0 = np.dot(new_data.iloc[i],gamma_hat)
        predictions.append(integral_fix + integral0 *lev_g[i]+sex.iloc[i]*theta_hat)       
    return np.array(predictions)
```

```python
fitted=compute_predictions(np.transpose(x))
```

```python
logit_fitted = scipy.special.expit(fitted)
labels = np.zeros(len(fitted))
ones= np.where(logit_fitted >0.5)[0]
labels[ones] = np.ones(len(ones)) 
acc = np.sum(labels==y23)/len(y23)
```
