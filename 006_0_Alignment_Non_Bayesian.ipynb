```python
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import fdasrsf as fs
import numpy as np
import json
```

```python
df = scipy.io.loadmat('TremorStruct_V1_Joint_Severity_Constancy_After_Exclusion.mat')
```

```python
var = df['TremorStruct'][0]
pat_number = len(var)
pat_number
```

```python
data = {}
for i in range(pat_number):
    data_pat = var[i]
    data[str(i)]=data_pat[17][0,0][0][:,:]
```

```python
data.keys()
```

```python
aligned = {}
for pat in data.keys():
    time = np.arange(data[pat].shape[0]).astype(np.float64)
    if len(time)>0:
        obj = fs.fdawarp(data[pat].astype(np.float64), time)
        obj.srsf_align(parallel=True)
        aligned[pat] = obj.fn
```

```python
aligned.keys()
```

```python
import pickle

with open('aligned_curves.pickle', 'wb') as handle:
    pickle.dump(aligned, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
